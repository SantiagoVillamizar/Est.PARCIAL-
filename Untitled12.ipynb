# TABLAS DE CONTIENGENCIA

```python
df2.loc[(df2.IDHEMOCLAS == 1), "IDHEMOCLAS"], df2.loc[(df2.IDHEMOCLAS == 2), "IDHEMOCLAS"], df2.loc[(df2.IDHEMOCLAS == 3), "IDHEMOCLAS"], df2.loc[(df2.IDHEMOCLAS == 4), "IDHEMOCLAS"] = "A","AB","B", "O"

df2.loc[(df2.IDFACTORRH == 1), "IDFACTORRH"], df2.loc[(df2.IDFACTORRH == 2), "IDFACTORRH"] = "Positivo","Negativo",
df2.head(3)
```

```python
tabla_cont = pd.crosstab(df2.IDHEMOCLAS, df2["IDFACTORRH"])
tabla_cont
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,6))
heatmap = sns.heatmap(tabla_cont, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.show()
```

```python
import scipy.stats as stats

res = stats.chi2_contingency(tabla_cont)
chi, valor_p, _, _ = res
print(f"El valor de la estadística Chi-cuadrado es {chi:.4f} y el valor p es {valor_p:.4f}")
```

```python
import math

# Total de observaciones
n = len(df2.IDFACTORRH)

# Coeficiente de contingencia
coeficiente_contingencia = math.sqrt(chi / (chi + n))

# Coeficiente Phi (solo válido para tablas 2x2, se incluye aquí como lo pediste)
coeficiente_phi = math.sqrt(chi / n)

# Coeficiente V de Cramer
k = min(tabla_cont.shape)  # filas o columnas mínimas
coeficiente_v_cramer = math.sqrt(chi / (n * (k - 1)))

print(f"Los coeficientes de contingencia, Phi y V de Cramer están respectivamente dados por: "
      f"{coeficiente_contingencia:.4f}, {coeficiente_phi:.4f}, {coeficiente_v_cramer:.4f}")
```
